```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
energy = pd.read_csv("energy (2).csv")
```

```python
energy
```

```python
albania = energy[energy['Entity'] == 'Albania']
romania = energy[energy['Entity'] == 'Romania']
bulgaria = energy[energy['Entity'] == 'Bulgaria']
kosovo = energy[energy['Entity'] == 'Kosovo']
slovenia = energy[energy['Entity'] == 'Slovenia']
croatia = energy[energy['Entity'] == 'Croatia']
greece = energy[energy['Entity'] == 'Greece']
serbiaandmontenegro = energy[energy['Entity'] == 'Serbia and Montenegro']
northmacedonia = energy[energy['Entity'] == 'Northern Macedonia']
bosniaandherzegovina = energy[energy['Entity'] == 'Bosnia and Herzegovina']
montenegro= energy[energy['Entity'] == 'Montenegro']
turkey=energy[energy['Entity'] == 'Turkey']

plt.figure(figsize = (25,10))

plt.scatter(albania['Year'], albania['Primary energy consumption per capita (kWh/person)'], color="brown", label="Albania")
plt.plot(albania['Year'], albania['Primary energy consumption per capita (kWh/person)'], color="brown")

plt.scatter(romania['Year'], romania['Primary energy consumption per capita (kWh/person)'], color="pink", label="Romania")
plt.plot(romania['Year'], romania['Primary energy consumption per capita (kWh/person)'], color="pink")

plt.scatter(bulgaria['Year'], bulgaria['Primary energy consumption per capita (kWh/person)'], color="seagreen", label="Bulgaria")
plt.plot(bulgaria['Year'], bulgaria['Primary energy consumption per capita (kWh/person)'], color="seagreen")

plt.scatter(kosovo['Year'], kosovo['Primary energy consumption per capita (kWh/person)'], color="green", label="Kosovo")
plt.plot(kosovo['Year'], kosovo['Primary energy consumption per capita (kWh/person)'], color="green")

plt.scatter(slovenia['Year'], slovenia['Primary energy consumption per capita (kWh/person)'], color="purple", label="Slovenia")
plt.plot(slovenia['Year'], slovenia['Primary energy consumption per capita (kWh/person)'], color="purple")

plt.scatter(croatia['Year'], croatia['Primary energy consumption per capita (kWh/person)'], color="rebeccapurple", label="Croatia")
plt.plot(croatia['Year'], croatia['Primary energy consumption per capita (kWh/person)'], color="rebeccapurple")

plt.scatter(greece['Year'], greece['Primary energy consumption per capita (kWh/person)'], color="tan", label="Greece")
plt.plot(greece['Year'], greece['Primary energy consumption per capita (kWh/person)'], color="tan")

plt.scatter(serbiaandmontenegro['Year'], serbiaandmontenegro['Primary energy consumption per capita (kWh/person)'], color="olive", label="Serbia and Montenegro")
plt.plot(serbiaandmontenegro['Year'], serbiaandmontenegro['Primary energy consumption per capita (kWh/person)'], color="olive")

plt.scatter( bosniaandherzegovina['Year'], bosniaandherzegovina['Primary energy consumption per capita (kWh/person)'], color="darkgreen", label="Bosnia and Herzegovina")
plt.plot( bosniaandherzegovina['Year'], bosniaandherzegovina['Primary energy consumption per capita (kWh/person)'], color="darkgreen")

plt.scatter( northmacedonia['Year'], northmacedonia['Primary energy consumption per capita (kWh/person)'], color="indianred", label="North Macedonia")
plt.plot( northmacedonia['Year'], northmacedonia['Primary energy consumption per capita (kWh/person)'], color="indianred")

plt.scatter( turkey['Year'], turkey['Primary energy consumption per capita (kWh/person)'], color="indianred", label="Turkey")
plt.plot( turkey['Year'], turkey['Primary energy consumption per capita (kWh/person)'], color="indianred")

plt.scatter( montenegro['Year'], montenegro['Primary energy consumption per capita (kWh/person)'], color="lightcoral", label="Montenegro")
plt.plot( montenegro['Year'], montenegro['Primary energy consumption per capita (kWh/person)'], color="lightcoral")

plt.xlabel('Year', fontsize=18)
plt.ylabel('Primary energy consumption per capita (kWh/person)', fontsize=18)
plt.title('Primary energy consumption per capita in the Balkan', fontsize=18)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=18)
plt.show()
```
